```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data (age of a car vs. its speed)
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6]).reshape((-1, 1))
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

# Display the data shape to confirm dimensions
print(f"Shape of x: {x.shape}")
print(f"Shape of y: {y.shape}")

model = LinearRegression()
model.fit(x, y)

r_sq = model.score(x, y)
print(f"Coefficient of determination (R-squared): {r_sq:.4f}")


# Get the intercept (b0) and coefficients (b1)
print(f"Intercept: {model.intercept_:.4f}")
print(f"Slope/Coefficient: {model.coef_[0]:.4f}")


# Predict response using the existing input data
y_pred = model.predict(x)
print(f"Predicted responses:\n{y_pred}")

# Predict response for new data (e.g., a 10-year-old car)
x_new = np.array([10]).reshape((-1, 1))
y_new = model.predict(x_new)
print(f"\nPrediction for a 10-year-old car: {y_new[0]:.4f}")

plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, y_pred, color='red', linewidth=2, label='Regression Line')
plt.xlabel('Age of Car (x)')
plt.ylabel('Speed (y)')
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```
